# 1. Introduction

"Generative Adversarial Networks is the most interesting idea in the last 10 years in Machine Learning." — Yann LeCun, 	
Turing Award (2018), founding father of convolutional nets.

## Short description
Generative Adversarial Networks (GAN) got an significant attention last few years and became a popular Machine Learning reshearch topic. Introduced by Ian Goodfellow to the delegates of the Neural Information Processing Systems (NIPS) conference in 2014, its tutorial entitled "Generative Adversarial Networks" have spawned a wide variety of variations on his core idea as Conditional GAN, Wasserstein GAN, Cycle GAN and so on.


GAN are composed of two models:
1. The Generator aims to generate new data as close as possible to the expected one from noise inputs. It could be seen as a student trying to recreate a task.
2. The Discriminator seek to distinguish fake or real data. The fake data is the generator outputs and the real ones belong to the original dataset. This model could be seen as a teacher wanting to detect if it's real or fake data. 

To do so, during training the generator tries to learn how to create data in such a way that the discriminator isn’t able to distinguish it as fake anymore. Due to this competition (adversarial), the generator improve its knowlegde and succeeds creating realistic data.

<img src="img/gan_scheme.png" alt="Drawing" style="width: 800px;"/>


## Mathematical context

Assume G(.) and D(.) be the generator and discriminator functions respectively. Both takes images as inputs.
We enter in a two agents minimax game with value function $V(G,D)$:  
$$\min_{G} \max_{D} V(G,D) = \mathbb{E}_{x\sim p_{data}(x)} [log(D(x))] +\mathbb{E}_{z\sim p_{z}(z)} [log(1-D(G(z)))]$$  
with $x$, $z$, real and fake samples respectively and $p_.(.)$ the distribution function.

### Binary Cross Entopy Loss
$L = {l_1,...l_N}^T$, with $l_i = -w_i[y_i log(v_i) + (1-y)log(1-v_i)]$  
This loss compares the discriminator's predictions on real date to an array of ones, and those on fake data (from generated predictions) to an array of zeroes.

### Generator
<img src="img/generator.png" alt="Drawing" style="width: 300px;"/>  

The Generator loss is defined as below:  
$\nabla_{\theta_G} \frac{1}{m} \sum_{i=1}^{m} log(1-D(G(z^{(i)})))$    , with $m$ the number of samples and $z$ a random noise.  
$D(G(z^{(i)}))$ determines how realistic are the generated samples, thus the generator aims to maximize it.  

### Discriminator
<img src="img/discriminator.png" alt="Drawing" style="width: 300px;"/>

The Discriminator loss is divided in two parts:  
$\nabla_{\theta_D} \frac{1}{m}\sum_{i=1}^m[log(D(x^{(i)})+log(1-D(G(z^{(i)})))]$  
The first term in the sum $logD(x^{(i)})$ assures that real samples are classified as being real, while the second term $log(1-D(G(z^{(i)})))$ make sure that generated samples are classified as fake.

# Wasserstein GAN

One way to improve the training of GANs is to use a better metric of the distribution similarity.  

The Jensen–Shannon divergence used as loss function in our vanilla GAN fails to provide a meaningful value when two distributions are disjoint. Wasserstein metric is proposed to replace JS divergence due to its smoother value space.  

## What is Wasserstein Distance?

Wasserstein Distance (which should be called Kantorovich–Rubinstein Distance) measures the distance between two probability distributions. Coming from Optimal Transport field, this metric can be interpreted as the minimum cost of moving and transforming a pile of dirt in the shape of one probability distribution to the shape of the other distribution. The cost is quantified as the amount of dirt moved times the moving distance.

$$W(p_\alpha,p_\beta) = \inf_{\gamma\sim\Pi(p_\alpha,p_\beta)}\mathbb{E}_{(x,y)\sim\gamma}\big[||x−y||\big]$$


## Mathematical context

In order to use Wasserstein distance as a GAN loss function, the authors of the original WGAN paper proposed a smart transformation of the formula based on the Kantorovich-Rubinstein duality:

$$W(p_\alpha,p_\beta) = \frac{1}{K} \sup_{||f||_L\leq K} \mathbb{E}_{x\sim p_\alpha}\big[f(x)\big] − \mathbb{E}_{x\sim p_\beta}\big[f(x)\big]$$
Thus, we want to maximize this value. Without going into details, it forces $f$ to be K-Lipschitz continuous.  
To fix it, the original WGAN paper applies a clipping to restrict the maximum weight value in $f$, i.e. the weights of the discriminator must be within a certain range controlled by an hyperparameters c.

$$W(p_\alpha,p_\beta) = \max_{w\in W} \mathbb{E}_{x\sim p_{data}}\big[D_w(x)\big] − \mathbb{E}_{z\sim p_z}\big[D_w(G(z))\big]$$
with $x$, $z$, real and fake samples respectively and $p_.(.)$ the distribution function.

The Generator loss is defined as below:  
$\nabla_{\theta_G} \frac{1}{m} \sum_{i=1}^{m} D(G(z^{(i)}))$    , with $m$ the number of samples and $z$ a random noise.      

The Discriminator loss is now:  
$\nabla_{\theta_D} \frac{1}{m}\sum_{i=1}^m[D(x^{(i)}) - D(G(z^{(i)}))]$  
The discriminator is no longer real or fake data critic but helps for estimating the Wasserstein metric in-between real and fake data distributions.

As mentioned above Wasserstein distance can still provide a meaningful and smooth representation of the distance between two disjoint distributions.  
However, Wasserstein GAN suffers from slow and unstable training. The authors of the original WGAN paper mentioned that "Weight clipping is a clearly terrible way to enforce a Lipschitz constraint". When clipping window is too large convergence can be significantly slow, and when clipping window is too small WGAN undergoes of vanishing gradients.
